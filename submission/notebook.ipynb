## Menyiapkan semua library yang dibuthkan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
# Membaca data air quality berformat csv untuk daerah Dongsi dalam bentuk data tabular
aqi_df = pd.read_csv("data/dataset.csv")
aqi_df.head(30)
```

### Assessing Data

```python
# Melakukan pemeriksaan data
aqi_df.info()
```

```python
# Mencari tahu nilai unik dalam kolom wd
aqi_df['wd'].nunique()
```

### Cleaning Data

```python
# mencari data kosong
aqi_df.isna().sum()
```

```python
# Mencari data yang  duplikat
print("Jumlah duplikasi: ", aqi_df.duplicated().sum())
```

```python
aqi_df.describe()
```

```python
# Mengapus data yang kosong dengan di drop
aqi_df.dropna(inplace=True)
```

Hasilnya tidak ada data yang kosong lagi

```python
aqi_df.isna().sum()
```

```python
# Melakukan reset index
aqi_df.reset_index(drop=True, inplace=True)
```

```python
aqi_df
```

Data duplikat dan kosong sudah hilang

```python
# Menyatukan kolom year, month, dan day menjadi satu kesatuan dengan nama kolom baru datetime
aqi_df["datetime"] = pd.to_datetime(aqi_df[['year', 'month', 'day']])
```

```python
aqi_df
```

```python
aqi_df.to_csv('Dataset_bersih.csv')
```

## Exploratory Data Analysis (EDA)

### Explore ...

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Memeriksa distribusi data
aqi_df.hist()
```

```python
# Memeriksa korelasi
aqi_df.corr()
```

```python
# Membuat pivot table dengan parameter year untuk mencari mean PM2.5 tertinggi
aqi_pertanyaan1 = aqi_df.groupby(by=aqi_df['year']).agg({
    'PM2.5': 'mean'
})
```

```python
aqi_pertanyaan1
```

```python
#Melakukan reset index agar kolom month dapat di visualisasi
aqi_pertanyaan1.reset_index(inplace=True)
```

```python
aqi_pertanyaan1
```

```python
# Membuat visualisasi grafik rata-rata PM2.5 dari tahun ke tahun
plt.plot(aqi_pertanyaan1['year'],aqi_pertanyaan1['PM2.5'])
plt.xlabel('Year')
plt.ylabel('Rata-rata PM2.5')
plt.show()
```

### Pertanyaan 2:

```python
# Membuat pivot table dengan parameter year untuk mencari O3 tertinggi
aqi_pertanyaan2 = aqi_df.groupby(by=aqi_df['year']).agg({
    'O3': 'max'
})
```

```python
aqi_pertanyaan2
```

```python
#Melakukan reset index agar kolom month dapat di visualisasi
aqi_pertanyaan2.reset_index(inplace=True)
```

```python
aqi_pertanyaan2
```

```python
# Membuat visualisasi grafik nilai tertinggi O3 dari tahun ke tahun
plt.plot(aqi_pertanyaan2['year'],aqi_pertanyaan2['O3'])
plt.xlabel('Year')
plt.ylabel('O3 tertinggi')
plt.show()
```
